```python
# This function will take uploaded PDF files and extract all the text.

from PyPDF2 import PdfReader
from typing import List

def get_pdf_text(pdf_docs: List):
    text = ""
    for pdf in pdf_docs:
        pdf_reader = PdfReader(pdf)
        for page in pdf_reader.pages:
            text += page.extract_text() or ""
    return text
```

```python
!pip install langchain

from langchain.text_splitter import RecursiveCharacterTextSplitter

def get_text_chunks(text: str) -> list[str]:
    # Initialize the text splitter with desired chunk size and overlap
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=10000,
        chunk_overlap=1000
    )
    # Split the input text into overlapping chunks
    chunks = text_splitter.split_text(text)
    return chunks
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain.vectorstores import FAISS

def get_vector_store(chunks):
    # Initialize the embedding model
    embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001")  # type: ignore

    # Create the vector store from text chunks
    vector_store = FAISS.from_texts(chunks, embedding=embeddings)

    # Save the index locally
    vector_store.save_local("faiss_index")
    print("✅ Vector store saved to 'faiss_index' directory.")
```

```python
# 🔍 What This Does:
	# •	Creates a QA agent that answers based only on context (from PDF chunks).
	# •	Ensures it doesn’t hallucinate—returns "Answer not available..." when needed.

from langchain.chains.question_answering import load_qa_chain
from langchain.prompts import PromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI

def get_conversational_chain():
    prompt_template = """
    You are an intelligent PDF assistant. Answer the question as truthfully and detailed as possible using the provided context.
    If the answer is not available in the context, reply: "Answer not available in the context."

    Context:
    {context}

    Question:
    {question}

    Answer:
    """

    # Initialize Gemini LLM
    model = ChatGoogleGenerativeAI(
        model="models/chat-bison-001",  # Or another Gemini-supported chat model
        client=None,  # Use env-based credentials
        temperature=0.3
    )

    # Define the LangChain prompt
    prompt = PromptTemplate(
        template=prompt_template,
        input_variables=["context", "question"]
    )

    # Load a Stuff QA chain with the LLM and prompt
    chain = load_qa_chain(
        llm=model,
        chain_type="stuff",
        prompt=prompt
    )

    return chain
```

```python
from langchain.vectorstores import FAISS
from langchain_google_genai import GoogleGenerativeAIEmbeddings

def user_input(user_question):
    # Step 1: Load Google Gemini embeddings
    embeddings = GoogleGenerativeAIEmbeddings(
        model="models/embedding-001"
    )

    # Step 2: Load FAISS vector store (ensure 'faiss_index' folder exists)
    new_db = FAISS.load_local(
        "faiss_index", embeddings, allow_dangerous_deserialization=True
    )

    # Step 3: Search for relevant chunks using the user's question
    docs = new_db.similarity_search(user_question)

    # Step 4: Load conversational chain
    chain = get_conversational_chain()

    # Step 5: Prepare context and ask the question
    context = "\n".join([doc.page_content for doc in docs])

    response = chain({
        "input_documents": docs,
        "context": context,
        "question": user_question
    }, return_only_outputs=True)

    # Step 6: Return just the answer
    return response['output_text']
```

```python
import os
import csv

def save_user_info(name, phone, email):
    file_exists = os.path.isfile('user_info.csv')
    with open('user_info.csv', mode='a', newline='') as file:
        fieldnames = ['Name', 'Phone', 'Email']
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        if not file_exists:
            writer.writeheader()
        writer.writerow({'Name': name, 'Phone': phone, 'Email': email})

# 🔹 Example test:
save_user_info("Michael Scott", "9866007834", "m.scott@gmail.com")
print("✅ User info saved!")
```

```python
import streamlit as st

def main():
    st.set_page_config(
        page_title="Gemini PDF Chatbot",
        page_icon="📄",
        layout="wide"
    )

    # --- Sidebar: Upload PDFs ---
    with st.sidebar:
        st.title("📑 Menu")
        pdf_docs = st.file_uploader(
            "Upload your PDF Files and Click on the Submit & Process Button",
            accept_multiple_files=True
        )
        if st.button("Submit & Process") and pdf_docs:
            with st.spinner("Processing..."):
                raw_text = get_pdf_text(pdf_docs)
                text_chunks = get_text_chunks(raw_text)
                get_vector_store(text_chunks)
                st.success("✅ Done")

    # --- Main content area ---
    st.title("💬 Chat with PDF files using Gemini 🙋‍♂️")
    st.write("Welcome to the chat!")

    if st.sidebar.button("Clear Chat History"):
        clear_chat_history()

    # Initialize chat history
    if "messages" not in st.session_state:
        st.session_state.messages = [{"role": "assistant", "content": "Upload some PDFs and ask me a question"}]

    # Display past messages
    for message in st.session_state.messages:
        with st.chat_message(message["role"]):
            st.markdown(f"**{message['role'].capitalize()}:** {message['content']}")

    # Handle new user prompt
    if prompt := st.chat_input("Ask something about your PDFs..."):
        st.session_state.messages.append({"role": "user", "content": prompt})
        with st.chat_message("user"):
            st.markdown(f"**User:** {prompt}")

        if "call me" in prompt.lower():
            st.session_state.collecting_info = True

        if not any(m["role"] == "assistant" for m in st.session_state.messages[-2:]):
            with st.chat_message("assistant"):
                with st.spinner("Thinking..."):
                    response = user_input(prompt)
                    st.session_state.messages.append({"role": "assistant", "content": response})
                    st.markdown(f"**Assistant:** {response}")

    # Collect user contact info if requested
    if "collecting_info" in st.session_state and st.session_state.collecting_info:
        st.subheader("📇 Please share your contact details")
        with st.form(key="contact_form"):
            name = st.text_input("Name")
            phone = st.text_input("Phone")
            email = st.text_input("Email")
            submit_button = st.form_submit_button("Submit")

            if submit_button:
                save_user_info(name, phone, email)
                st.session_state.messages.append({
                    "role": "assistant",
                    "content": f"Thank you, {name}. We will contact you at {phone} or {email}."
                })
                st.session_state.collecting_info = False

# 🔹 Entry point
if __name__ == "__main__":
    main()
```
