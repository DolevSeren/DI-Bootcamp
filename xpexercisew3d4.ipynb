```python
import matplotlib.pyplot as plt

x = ["sunday", "monday", "tuesday", "wednesday", "thursday", "friday", "saturday"]
y = [72, 74, 76, 80, 82, 78, 75]

plt.plot(x, y)
plt.xlabel('day')
plt.ylabel('temperature')
plt.title("Temperature in a week")
plt.show()
```

```python
sales = [5000, 5500, 6200, 7000, 7500]
month = ["jan", "feb", "march" ,"apr", "may"]
plt.bar(month,sales)
plt.xlabel("sales")
plt.ylabel("month")
plt.title("sales in a month")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# דוגמה ליצירת df עם ערכים רנדומליים
# ניתן לדלג על זה אם כבר יש לך את df
import numpy as np
df = pd.DataFrame({'CGPA': np.random.normal(3.0, 0.5, 100)})

# יצירת ההיסטוגרמה
sns.histplot(data=df, x='CGPA', color='skyblue', bins=10)
plt.title('Distribution of Students\' CGPA')
plt.xlabel('CGPA')
plt.ylabel('Number of Students')
plt.show()
```

```python


# דוגמת נתונים
df = pd.DataFrame({
    'Do you have Anxiety?': ['Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'No', 'Yes'],
    'Choose your gender': ['Female', 'Female', 'Male', 'Female', 'Male', 'Male', 'Other', 'Female', 'Other']
})

# חישוב פרופורציות
anxiety_df = df[df['Do you have Anxiety?'] == 'Yes']
gender_counts = anxiety_df['Choose your gender'].value_counts()
total_counts = df['Choose your gender'].value_counts()
proportions = (gender_counts / total_counts).reset_index()
proportions.columns = ['Gender', 'Proportion']

# ציור הגרף
sns.barplot(data=proportions, x='Gender', y='Proportion', palette='pastel')
plt.title('Proportion of Students with Anxiety by Gender')
plt.xlabel('Gender')
plt.ylabel('Proportion with Anxiety')
plt.ylim(0, 1)
plt.show()
```

```python

df['PanicNumeric'] = df['Do you have Panic Attacks?'].map({'Yes': 1, 'No': 0})
sns.scatterplot(data=df, x='Age', y='PanicNumeric', alpha=0.6, s=80, color='tomato')
plt.title('Relationship Between Age and Panic Attacks')
plt.xlabel('Age')
plt.ylabel('Panic Attack (1=Yes, 0=No)')
plt.yticks([0, 1], ['No', 'Yes'])  # שיהיה ברור וקריא

```
