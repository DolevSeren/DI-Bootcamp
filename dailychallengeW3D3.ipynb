```python
import pandas as pd

df = pd.read_csv("/content/sample_data/datascience_salaries.csv")

```

```python
print(df.columns)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['Salary_normalized'] = scaler.fit_transform(df[['salary']])
```

```python
df.head()
```

```python
from sklearn.decomposition import PCA

numeric_cols = df.select_dtypes(include=['float64', 'int64'])
scaled_data = MinMaxScaler().fit_transform(numeric_cols)
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(scaled_data)
pca_df = pd.DataFrame(reduced_data, columns=['PC1', 'PC2'])

```

```python
salary_stats = df.groupby('experience_level')['salary'].agg(['mean', 'median']).reset_index()
print(salary_stats)
```
