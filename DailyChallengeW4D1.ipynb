```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
matrix = np.random.uniform(-5, 35 ,size=(10,12))
```

```python
cities = [f'City {i+1}' for i in range(10)]
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
df = pd.DataFrame(matrix, index=cities, columns=months)
```

```python
df.head()
```

```python
annual_avg = df.mean(axis=1)
```

```python
hottest_city = annual_avg.idxmax()
coldest_city = annual_avg.idxmin()

print("City with highest average temp:", hottest_city)
print("City with lowest average temp:", coldest_city)
```

```python
plt.figure(figsize=(12, 6))

for city in df.index:
  plt.plot(df.columns, df.loc[city], label=city)

plt.title("Monthly Temperature Trends by City")
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.legend(loc="center left", bbox_to_anchor=(1, 0.5))
plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```
