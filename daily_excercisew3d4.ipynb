```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("/2019.csv")
df.head()

```

```python
missing_counts = df.isnull().sum()
print(missing_counts)
```

```python
sns.lmplot(x='Social support', y='Score', data=df)
plt.show()
```

```python
selcted_country = ["Israel", "Finland", "Canada", "Japan", "Brazil"]
filtrized_df = df[df["Country or region"].isin(selcted_country)]

sns.lmplot(x='GDP per capita', y='Healthy life expectancy', data=filtrized_df, hue="Country or region")
plt.show()
```
