```python
mean = df5['PetalLengthCm'].mean()
median = df5['SepalLengthCm'].median()
mode = df5['PetalLengthCm'].mode()[0]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.scatter(df5['PetalLengthCm'], df5['SepalLengthCm'], alpha=0.7)
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Sepal Length (cm)')
plt.grid(True)
plt.show()


```
